### 1. Determinar las variables que rigen el sistema.

El Juego de la Vida, se basa en un conjunto de celdas en las cuales pueden o no haber células (celdas pintadas), en base a determinadas reglas. En ese contexto, el sistema establece cuatro reglas dependiendo si una celda está o no poblada.

**Si la celda está poblada:**
1.   **Soledad:** si una célula tiene una célula vecina, o no tiene ninguna, muere.
2.   **Sobrepoblación:** si una célula tiene más de tres células vecinas, muere.
3.   **Supervivencia:** si una célula tiene exactamente dos o tres células vecinas, permanece.

**Si la celda NO está poblada:**
4.   **Reproducción:** si un espacio vacío tiene como vecinas a exactamente tres células, dentro de ese espacio nace una célula.

En resumen, el comportamiento se define por dos cosas: la cantidad de vecinos de una célula, y el tiempo que transcurra desde la ejecución del sistema (las generaciones). 

### 2. Diseñar un plan de simulación.

Dentro del plan de simulación se deben establecer configuraciones para que las células no varíen. Por lo tanto, hay que tener en cuenta la regla de **Supervivencia**, para asegurarse que las células no mueran, y la regla de **Reproducción**, para evitar que nuevas células sean creadas.

Teniendo esto en cuenta, es necesario "diseñar" una combinación de células iniciales que tengan dos o tres células vecinas, y que los espacios vacíos entre células, y sus alrededores, no tengan exactamente tres células. Al encontrar patrones que cumplan con lo mencionado, se asegurará que las células sean inmutables sin importar cuántas generaciones pasen (tiempo transcurrido).
