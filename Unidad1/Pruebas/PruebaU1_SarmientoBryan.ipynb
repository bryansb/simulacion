```python
def add_value_label(x_list,y_list):
    for i in range(0, len(x_list)):
        plt.text(i,y_list[i],y_list[i], ha="center")
```

```python
xlsx_file = io.BytesIO(uploaded.get('Prueba.xlsx'))
df_general = pd.read_excel(xlsx_file, 'RESUMEN DE GRUPOS ECON√ìMICO', engine='openpyxl')
df_panama = pd.read_excel(xlsx_file, 'PANAM√Å PAPERS', engine='openpyxl')
```

```python
# Gr√°fico 1
better = df_general.loc[df_general['Ranking \n(1) Tama√±o  2021'] > df_general['Ranking \n(1) Tama√±o  2020']].count()[0]
worse = df_general.loc[df_general['Ranking \n(1) Tama√±o  2021'] < df_general['Ranking \n(1) Tama√±o  2020']].count()[0]
# print(better, worse)

# Par√°metros para la gr√°fica
labels = ['Empeoraron', 'Mejoraron']
values = [worse, better]

plt.figure(dpi=300)
plt.bar(labels, values)
add_value_label(labels, values)
plt.title('Cambios en el ranking')
plt.xlabel('Empresas')
plt.ylabel('N√∫mero de empresas')
plt.show()
```

```python
#Gr√°fico 2
num_panama = df_general.loc[df_general['No. de Integrantes y Empresas Offshore relacionadas con el Grupo Econ√≥mico publicados en PANAM√Å PAPERS (***)'] != 0].count()[0]
num_no_panama = df_general.loc[df_general['No. de Integrantes y Empresas Offshore relacionadas con el Grupo Econ√≥mico publicados en PANAM√Å PAPERS (***)'] == 0].count()[0]
# print(num_panama, num_no_panama)

# Par√°metros para la gr√°fica
labels = ['Con Panam√° Papers', 'Sin Panam√° Papers']
values = [num_panama, num_no_panama]

plt.figure(dpi=300)
plt.bar(labels, values)
add_value_label(labels, values)
plt.title('Empresas con Panam√° Papers')
plt.xlabel('Empresas')
plt.ylabel('N√∫mero de empresas')
plt.show()
```

```python
#Gr√°fico 3
aporte = df_general.sort_values(by=['\nTotal Recaudado (8) 2020'], ascending=False)
aporte_final = aporte.iloc[0:10]
# aporte_final

plt.figure(dpi=300)
plt.pie(aporte_final['\nTotal Recaudado (8) 2020'], 
            autopct='%1.2f %%')
plt.title('Empresas con mayor aporte')
plt.legend(aporte_final['\nNombre Grupo Econ√≥mico'], 
           title = 'Rangos', 
           loc = 'upper', 
           bbox_to_anchor = (1, 0, 0.5, 1))
plt.show()
```

```python
#Gr√°fico 4
max = df_general.sort_values(by=['Total Patrimonio (7) 2020'], ascending=False)
max_final = max.iloc[0:10]
# max_final

# Se crea y presenta la figura
plt.figure(dpi=300)
plt.pie(max_final['Total Patrimonio (7) 2020'], 
            autopct='%1.2f %%')
plt.title('Empresas con mayor patrimonio')
plt.legend(max_final['\nNombre Grupo Econ√≥mico'], 
           title = 'Rangos', 
           loc = 'upper', 
           bbox_to_anchor = (1, 0, 0.5, 1))
plt.show()
```
