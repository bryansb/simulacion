## Análisis descriptivo - FIFA

```python
path = './fifa_datos.csv'

data_fifa = pd.read_csv(path, header=0, delimiter=',')
```

```python
data_fifa = data_fifa.drop(['Unnamed: 0'], axis=1)
data_fifa.head()
```

```python
# Información General del Dataset
data_fifa.info()
```

```python
# Descripción numérica: int64
data_fifa.describe(include=['int64'])
```

```python
# Descripción numérica: float64
data_fifa.describe(include=['float64'])
```

```python
# Descripción categórica
data_fifa.describe(include=[object])
```

```python
sns.heatmap(data_fifa.corr())
```
