### 1. Determinar las variables que rigen el sistema.

El Juego de la Vida, se basa en un conjunto de celdas en las cuales pueden, o no, haber células (celdas pintadas), en base a determinadas reglas. En ese contexto, el sistema establece una serie de reglas dependiendo de la situación.

**Si la celda está poblada:**
*   **Soledad:** si una célula tiene una célula vecina, o no tiene ninguna, muere.
*   **Sobrepoblación:** si una célula tiene más de tres células vecinas, muere.
*   **Supervivencia:** si una célula tiene exactamente dos o tres células vecinas, permanece.

**Si la celda NO está poblada:**
*   **Reproducción:** si un espacio vacío tiene como vecinas a exactamente tres células, dentro de ese espacio nace una célula.

En resumen, el comportamiento se define por dos cosas: la cantidad de vecinos de una célula, y el tiempo que transcurra desde la ejecución del sistema (las generaciones). 

### 2. Diseñar un plan de simulación.

#### 2.1. Plan para lograr que no varíen las células
Para lograr esto, hay que tener en cuenta la regla de **Supervivencia**, para asegurarse que las células no mueran, y la regla de **Reproducción**, para evitar que nuevas células sean creadas.

Teniendo esto en cuenta, es necesario "diseñar" una combinación de células iniciales que tengan dos o tres células vecinas, y que los espacios vacíos entre células, y sus alrededores, no tengan exactamente tres células. Al encontrar patrones que cumplan con lo mencionado, se asegurará que las células sean inmutables sin importar cuántas generaciones pasen (tiempo transcurrido).

#### 2.2. Plan para lograr un crecimiento de células

Para lograr esto, es necesario que la regla de **Reproducción** se cumpla, contrario al plan anterior. Por lo tanto, es necesario crear patrones en los cuales haya espacios vacíos con tres vecinos celulares.

Además, se debería evitar excesivamente la **Sobrepoblación** y la **Soledad**, para evitar que lleguen a extinguirse completamente las células al pasar determinadas generaciones.
